**Import libraries**

```python
#import pandas and sklearn - from the past colab
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

**Read file**

```python
# Use pd.read_excel to read excel file
gymReco = pd.read_excel('/content/sample_data/gym-recommendation.xlsx')
gymReco

# Gender = male = 1 female = 0

# Fitness Type = Muscular fitness = 0 Cardiovascular fitness = 1
```

**Seperate X and y columns**

```python
# Create variables each for X and y , then subset the gym data
# Input must have the gender, bmi, and fitness type columns, and output is the exercise column
x = gymReco[['Gender', 'BMI', 'Fitness Type']]
y = gymReco['Exercises']

# x
y
```

**Split Data**

```python
#Set random state to 0 and the test size to 0.2 - copied from the past colab
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size = 0.2)
```

**Train the model**

```python
# Create a variable named model and choose an algorithm to train it, the fit the training data
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

**Test the model**

```python
# Create a variable named ytestpredictions and then predict the model using the X_test data
y_test_predictions = model.predict(x_test)
```

**Compare actual and predicted values**

```python
# Create a variable named predictions and create a DataFrame of the actual and predictions data using y_test and ytestpredictions
# join the X_test with the predictions DataFrame and show the created DataFrame
predictions = pd.DataFrame({
    'actual': y_test,
    'prediction': y_test_predictions
}, index = y_test.index)

predictions = x_test.join(predictions)
predictions
```

**Check accuracy score**

```python
# Create a variable named score and then use y_test and ytestpredictions to check the accuracy score, then show the score
score = accuracy_score(y_test,y_test_predictions)
score
# if aint 41% low 😔
```
